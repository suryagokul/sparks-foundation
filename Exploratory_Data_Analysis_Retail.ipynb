### `Heatmap`

```
sns.heatmap(df.corr(),annot=True,cmap='spring')
```

### State Wise Business 

```
plt.figure(figsize=(15,6))

sns.countplot(df.State)

sns.countplot(df.State).set_xticklabels(sns.countplot(df.State).get_xticklabels(), rotation=90, ha="right")

fig = plt.gcf()
fig.set_size_inches(15,5)
plt.title('States')
```

###  Region Wise Business

```
plt.figure(figsize=(15,6))

sns.countplot(df.Region)

sns.countplot(df.Region).set_xticklabels(sns.countplot(df.Region).get_xticklabels(), rotation=90, ha="right")

fig = plt.gcf()
fig.set_size_inches(15,5)
```

### Number Of Categories in Store

```
plt.figure(figsize=(15,6))

sns.countplot(df.Category)

sns.countplot(df.Category).set_xticklabels(sns.countplot(df.Category).get_xticklabels(), rotation=90, ha="right")

fig = plt.gcf()
fig.set_size_inches(15,5)
plt.title('Categories')
```

`For this we can say that, more no of Office Supplies are there in superstore.`

### Number Of Sub Categories are there in Store

```
plt.figure(figsize=(15,6))

sns.countplot(df['Sub-Category'])

sns.countplot(df['Sub-Category']).set_xticklabels(sns.countplot(df['Sub-Category']).get_xticklabels(), rotation=90, ha="right")

fig = plt.gcf()
fig.set_size_inches(15,5)
plt.title('Sub Categories')
```

### Region Wise Discount

```
loc_plt=pd.crosstab(df['Region'],df.Discount)
loc_plt.plot(kind='bar',stacked=True);
plt.title('Region wise - Discount',fontsize=15,fontweight='bold')
plt.ylabel('Region',fontsize=10,fontweight='bold')
plt.xlabel('Disocunt',fontsize=10,fontweight='bold')
plt.xticks(fontsize=10,fontweight='bold')
plt.yticks(fontsize=10,fontweight='bold');
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

`In West and East,  60% of goods were doesn't have any discounts compared to other regions like Central and South.`

### Category Wise Profit

```
df.groupby('Category')['Profit'].sum().plot(kind='bar',cmap='spring')
 
fig = plt.gcf()      

# gcf() is primarily used to get the current figure. If no current figure is available then one is created with the help of the figure() function.

fig.set_size_inches(14.5, 5)

plt.title('Category Wise - Profit')
```

`FOR TECHNOLOGY, there is more profit than others like Furniture and Office Supplies`

### Region Wise Profit

```
df.groupby('Region')['Profit'].sum().plot(kind='bar',cmap='cividis')
 
fig = plt.gcf()      

# gcf() is primarily used to get the current figure. If no current figure is available then one is created with the help of the figure() function.

fig.set_size_inches(14.5, 5)

plt.title('Region Wise - Profit')
```

`By seeing this, We clearly say that Region Of West Got More Profit  and then East.In central region there is less profit.`

### Category Wise Sales

```
df.groupby('Category')['Sales'].sum().plot(kind='bar',cmap='cool')
 
fig = plt.gcf()      

# gcf() is primarily used to get the current figure. If no current figure is available then one is created with the help of the figure() function.

fig.set_size_inches(14.5, 5)

plt.title('Category Wise - Sales')
```

`By seeing Category wise profit and Category wise sales, Furniture has done higher number of sales than Office Supplies but Furniture goods has very less profit.`

### Sub Categories and Quantities sold

```
df.groupby('Sub-Category')['Quantity'].sum().plot(kind='bar',cmap='Wistia')

fig = plt.gcf()

fig.set_size_inches(14.5,5)

plt.title('Sub Categories and Quantities sold')
```
